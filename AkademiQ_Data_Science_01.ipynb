```python
# Eksik veri oranına bakmak için
missing_ratio = data.isnull().mean().sort_values(ascending=False)
print("\nEksik veri oranı")
print(missing_ratio)
# ascending ile fazla eksik veriden az olan eksik veriye sıralama yaptı
# önce isnull ile eksik verileri buluyor sonra mean ile ortalama alıyor sort ile sıraladı
```

**STANDARDİZASYON**

```python
data.columns = (data.columns
                .str.strip()
                .str.lower()
                .str.replace(" ", "_")
                .str.replace("-", "_"))
print("\nTemizlenmiş Sütun İsimleri:")
print(data.columns.tolist())
# ya da
print("\nTemizlenmiş Sütun İsimleri Liste:")
print("\n".join(data.columns))

# strip sütunlarının başındaki ve sonundaki boşlukları siler ör: "  Age  " -> "Age"
# strip metnin içindeki boşlukla ilgilenmiyor sadece başındaki ve sonundaki boşlukla ilgileniyor
# lower " Age " -> " age "
# replace boşluk yerine _ bunu koyar ör: "customer name" -> "customer_name"
```

```python
# Bu kod, veri setindeki metin (string/object) sütunlarını temizlemek için kullanılıyor.
string_cols = data.select_dtypes(include=['object'])
for i in string_cols:
  data[i] = data[i].astype(str).str.strip() # 123 -> "123", none-> "none" , " 123 " -> "123"
# Yani kısaca sütundaki verileri stringe dönüştürür ve boşlukları siler
# .str -> sadece string olan sütunlarda strip işlemini uygulamak için kullanılır
# bu işlemi yapmazsak hata alabiliriz "123" 123 " 123" " 123 " bunların hepsi farklı değerdir
```

```python
# Kritik 2 sütun ekleyip içinde eksik veri varsa silme yapma:
critical_cols = ["repository_name","stars_count"] # repo adını ve yıldız sayısını kritik sütun olarak seçtik
data = data.dropna(subset=critical_cols) # dropna -> drop sil demek na eksik veri demek
```

```python
numeric_cols = ["stars_count","forks_count","watchers_count","open_issues_count"]

for col in numeric_cols:
  if col in data.columns: # kod hata vermesin diye koşul verdik koşul diyor ki eğer sütun ismi bu listede varsa
    data[col] = pd.to_numeric(data[col], errors='coerce') # verileri numeric değere çevirir "123" -> 123  "50.5" -> 50.5

# fakat "abc" gibi değeri numerice çeviremez  errors=coerce : "abc"-> NaN (eksik veri)
#NOT
# NaN = Not a Number (Sayısal değer yok der. numpy ile gelir. Kullanmak için numpy'a ihtiyaç vardır.)
# NA = Not Available (genel olarak (tür belirtmeden) veri yok demektir)
```
